```python
# ===============================================
# 🔹 DETECCIÓN DE INTRUSIONES CON RED NEURONAL (NSL-KDD)
# ===============================================

# 1️⃣ Importar librerías
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

# 2️⃣ Cargar los datos
train_df = pd.read_csv("/content/train.csv")
test_df = pd.read_csv("/content/test.csv")

print("Tamaño del conjunto de entrenamiento:", train_df.shape)
print("Tamaño del conjunto de prueba:", test_df.shape)
print("\nColumnas del dataset:\n", list(train_df.columns))

# 3️⃣ Verificar que exista la columna 'class'
if 'class' not in train_df.columns:
    raise KeyError("❌ No se encontró la columna 'class'. Verifica el nombre en train_df.columns")

# 4️⃣ Codificar las columnas categóricas
categorical_cols = ['protocol_type', 'service', 'flag']
le = LabelEncoder()

for col in categorical_cols:
    train_df[col] = le.fit_transform(train_df[col])
    test_df[col] = le.transform(test_df[col])

# 5️⃣ Convertir la etiqueta 'class' en binaria (0 = normal, 1 = ataque)
train_df['class'] = train_df['class'].apply(lambda x: 0 if x == 'normal' else 1)
test_df['class'] = test_df['class'].apply(lambda x: 0 if x == 'normal' else 1)

# 6️⃣ Normalizar los datos numéricos
scaler = MinMaxScaler()

# MODIFICACIÓN: Manejar valores NaN antes de dividir y escalar
# Verifica y elimina filas con NaNs en ambos dataframes
if train_df.isnull().sum().sum() > 0:
    print("⚠️ ¡NaNs detectados en train_df! Eliminando filas con NaNs.")
    train_df = train_df.dropna()
if test_df.isnull().sum().sum() > 0:
    print("⚠️ ¡NaNs detectados en test_df! Eliminando filas con NaNs.")
    test_df = test_df.dropna()

X_train = train_df.drop(['class'], axis=1)
y_train = train_df['class']
X_test = test_df.drop(['class'], axis=1)
y_test = test_df['class']

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("\n✅ Datos listos para el entrenamiento.")

# 7️⃣ Crear el modelo neuronal (MLP)
model = Sequential([
    Dense(128, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

# 8️⃣ Entrenar el modelo
history = model.fit(
    X_train_scaled, y_train,
    epochs=20,
    batch_size=128,
    validation_split=0.2,
    verbose=1
)

# 9️⃣ Evaluar el modelo
y_pred = (model.predict(X_test_scaled) > 0.5).astype("int32")

acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("\n📊 --- RESULTADOS ---")
print(f"Accuracy:  {acc:.4f}")
print(f"Precision: {prec:.4f}")
print(f"Recall:    {rec:.4f}")
print(f"F1-Score:  {f1:.4f}")

# 🔟 Matriz de confusión
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Normal', 'Ataque'])
disp.plot(cmap='Purples')
plt.title("Matriz de Confusión")
plt.show()

# 🔢 Gráfico de desempeño
plt.plot(history.history['accuracy'], label='Entrenamiento')
plt.plot(history.history['val_accuracy'], label='Validación')
plt.title('Precisión del Modelo')
plt.xlabel('Época')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()







# ===============================================
# 🔹 MODELOS CLÁSICOS PARA COMPARACIÓN
# ===============================================

from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC

print("\n==============================")
print("🏆 COMPARACIÓN CON MODELOS CLÁSICOS")
print("==============================\n")

# 1️⃣ Decision Tree
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train_scaled, y_train)

dt_pred = dt_model.predict(X_test_scaled)

dt_acc = accuracy_score(y_test, dt_pred)
dt_prec = precision_score(y_test, dt_pred)
dt_rec = recall_score(y_test, dt_pred)
dt_f1 = f1_score(y_test, dt_pred)

print("🌲 Decision Tree")
print(f"Accuracy:  {dt_acc:.4f}")
print(f"Precision: {dt_prec:.4f}")
print(f"Recall:    {dt_rec:.4f}")
print(f"F1-Score:  {dt_f1:.4f}\n")

# 2️⃣ Support Vector Machine
# Nota: Para acelerar SVM, usamos kernel lineal
svm_model = SVC(kernel='linear')
svm_model.fit(X_train_scaled, y_train)

svm_pred = svm_model.predict(X_test_scaled)

svm_acc = accuracy_score(y_test, svm_pred)
svm_prec = precision_score(y_test, svm_pred)
svm_rec = recall_score(y_test, svm_pred)
svm_f1 = f1_score(y_test, svm_pred)

print("⚙️ Support Vector Machine (SVM)")
print(f"Accuracy:  {svm_acc:.4f}")
print(f"Precision: {svm_prec:.4f}")
print(f"Recall:    {svm_rec:.4f}")
print(f"F1-Score:  {svm_f1:.4f}\n")

# ===============================================
# 🔹 TABLA COMPARATIVA
# ===============================================

results = pd.DataFrame({
    'Modelo': ['MLP (Red Neuronal)', 'Decision Tree', 'SVM'],
    'Accuracy': [acc, dt_acc, svm_acc],
    'Precision': [prec, dt_prec, svm_prec],
    'Recall': [rec, dt_rec, svm_rec],
    'F1-Score': [f1, dt_f1, svm_f1]
})

print("\n📊 TABLA COMPARATIVA DE MÉTRICAS:\n")
print(results)

# ===============================================
# 🔹 GRÁFICO COMPARATIVO (Accuracy)
# ===============================================

plt.figure(figsize=(8,5))
plt.bar(results['Modelo'], results['Accuracy'], color=['purple','orange','blue'])
plt.title("Comparación de Accuracy entre Modelos")
plt.ylabel("Accuracy")
plt.ylim(0,1)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.show()
```
